```
import numpy as np
import pandas as pd

# Data file => https://online.stat.psu.edu/stat462/sites/onlinecourses.science.psu.edu.stat462/files/data/leukemia_remission/index.txt
column_names = ['user_id', 'movie_id', 'rating','timestamp']
data_df = pd.read_csv('/content/leukemia_remission.txt', sep='\t')
# REMISS	CELL	SMEAR	INFIL	LI	BLAST	TEMP

data_df.sample(5)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display_html 

# Generating Crosstabs for each feature wrt result (REMISS)
cell = pd.crosstab(data_df['CELL'], data_df['REMISS'])
smear = pd.crosstab(data_df['SMEAR'], data_df['REMISS'])
infil = pd.crosstab(data_df['INFIL'], data_df['REMISS'])
li = pd.crosstab(data_df['LI'], data_df['REMISS'])
blast = pd.crosstab(data_df['BLAST'], data_df['REMISS'])
temp = pd.crosstab(data_df['TEMP'], data_df['REMISS'])

# Printing all crosstabs in one row
df1_styler = cell.style.set_table_attributes("style='display:inline'").set_caption('CELL')
df2_styler = smear.style.set_table_attributes("style='display:inline'").set_caption('SMEAR')
df3_styler = infil.style.set_table_attributes("style='display:inline'").set_caption('INFIL')
df4_styler = li.style.set_table_attributes("style='display:inline'").set_caption('LI')
df5_styler = blast.style.set_table_attributes("style='display:inline'").set_caption('BLAST')
df6_styler = temp.style.set_table_attributes("style='display:inline'").set_caption('TEMP')
display_html(df1_styler._repr_html_()+df2_styler._repr_html_()+df3_styler._repr_html_()+df4_styler._repr_html_()+df5_styler._repr_html_()+df6_styler._repr_html_(), raw=True)

# Plotting Stacked Bar Plots for above crosstabs
stacked_cell = cell.stack().reset_index().rename(columns={0:'value'})
stacked_smear = smear.stack().reset_index().rename(columns={0:'value'})
stacked_infil = infil.stack().reset_index().rename(columns={0:'value'})
stacked_li = li.stack().reset_index().rename(columns={0:'value'})
stacked_blast = blast.stack().reset_index().rename(columns={0:'value'})
stacked_temp = temp.stack().reset_index().rename(columns={0:'value'})

# Creating a 2X3 Plot table
fig, axs = plt.subplots(2, 3, sharex=True, figsize=(30,10))

# Plot grouped bar chart
sns.barplot(x=stacked_cell.CELL, y=stacked_cell.value, hue=stacked_cell.REMISS, ax=axs[0][0])
sns.barplot(x=stacked_smear.SMEAR, y=stacked_smear.value, hue=stacked_smear.REMISS, ax=axs[0][1])
sns.barplot(x=stacked_infil.INFIL, y=stacked_infil.value, hue=stacked_infil.REMISS, ax=axs[0][2])
sns.barplot(x=stacked_li.LI, y=stacked_li.value, hue=stacked_li.REMISS, ax=axs[1][0])
sns.barplot(x=stacked_blast.BLAST, y=stacked_blast.value, hue=stacked_blast.REMISS, ax=axs[1][1])
sns.barplot(x=stacked_temp.TEMP, y=stacked_temp.value, hue=stacked_temp.REMISS, ax=axs[1][2])
```

```
# Dividing dataset into X Y sets
x = data_df.iloc[:, data_df.columns != 'REMISS']
y = data_df.iloc[:, 0]
print("X => \n", x.sample(5))
print("\nY => \n", y.sample(5))
```

```
# Correlation Analysis
correlation_matrix = x.corr().abs()
correlation_matrix
plt.subplots(figsize=(10,5))
sns.heatmap(correlation_matrix)
print("Correlation Matrix : \n", correlation_matrix)
print("Shape before dropping correlated columns : " , x.shape)
correlated_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) >= 0.5:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
print("Correlated Features : ", correlated_features)


# Dropping highly correlated features : 
# for col in correlated_features:
#   del x[col]

print("Shape after dropping correlated columns : " , x.shape)
```

```
# Data Prep
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Splitting into training & test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)

# evaluate a logistic regression model using k-fold cross-validation
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix

# Preparing for 10 Fold Cross Validation
cv = KFold(n_splits=10, random_state=1, shuffle=True)
# Creating Model
model = LogisticRegression()
# Evaluate Model
scores = cross_val_score(model, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# Report Performance
print('Cross Val Score Mean(Std): %.3f (%.3f)' % (mean(scores), std(scores)))

# Model Training & Results
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

f1_LR=f1_score(y_test,y_pred, average='macro')
cm = confusion_matrix(y_test, y_pred)
print("---------------------------------------------------------")
print("Confusion Matrix : \n", cm, "\n")
print("---------------------------------------------------------")
print("Classification Report : \n", classification_report(y_test,y_pred)) 
print("---------------------------------------------------------")
print("Training Accuracy: ", model.score(x_train, y_train))
print("Testing Accuracy: ", model.score(x_test, y_test))
print("---------------------------------------------------------")

print("\n==============================================================================================================")
# We know that Logit Function => 𝑓(𝑥₁, 𝑥₂) = 𝑏₀ + 𝑏₁𝑥₁ + 𝑏₂𝑥₂ + . . . .
print("Logit Function : 𝑓(𝑥₁, 𝑥₂, 𝑥₃, 𝑥₄, 𝑥₅, 𝑥₆) = 𝑏₀ + 𝑏₁𝑥₁ + 𝑏₂𝑥₂ + 𝑏₃𝑥₃ + 𝑏₄𝑥₄ + 𝑏₅𝑥₅ + 𝑏₆𝑥₆")
print("Logistic Regression Intercept (𝑏₀) : ", model.intercept_)
print("Logistic Regression Coeffecients (𝑏₁, 𝑏₂, 𝑏₃, 𝑏₄, 𝑏₅, 𝑏₆) : ", model.coef_)

logr_eqn = str(round(model.intercept_[0],3))
q = 1
for coef in model.coef_[0]:
  logr_eqn = logr_eqn + " + " + "(" + str(round(coef,3)) + ")" + "x" + str(q)
  q = q + 1
print("==============================================================================================================")
print("Logistic Regression Equation : ")
print("---------------------------------------------------------")
print("𝑓(𝑥₁, 𝑥₂, 𝑥₃, 𝑥₄, 𝑥₅, 𝑥₆) = ", logr_eqn)
print("==============================================================================================================")
```
